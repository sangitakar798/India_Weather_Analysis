```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
df = df.dropna(subset=['date'])
```

```python
df['rainfall'].fillna(0, inplace=True)
```

```python
df['max_humidity'].fillna(df['max_humidity'].median(), inplace=True)
df['min_humidity'].fillna(df['min_humidity'].median(), inplace=True)
df['max_temperature'] = df['max_temperature'].fillna(df['max_temperature'].mean())
df['min_temperature'] = df['min_temperature'].fillna(df['min_temperature'].mean())
df['wind_speed'] = df['wind_speed'].fillna(df['wind_speed'].mean())
df['pressure_surface_level'] = df['pressure_surface_level'].fillna(df['pressure_surface_level'].mean())
df['visibility'] = df['visibility'].fillna(df['visibility'].median())
df['cloud_cover'] = df['cloud_cover'].fillna(df['cloud_cover'].median())
df['solar_radiation'] = df['solar_radiation'].fillna(df['solar_radiation'].mean())
df['dew_point'] = df['dew_point'].fillna(df['dew_point'].mean())
df['uv_index'] = df['uv_index'].fillna(df['uv_index'].median())
df['heatwave'] = df['heatwave'].fillna(0)
df['precipitation_probability'] = df['precipitation_probability'].fillna(
    df['precipitation_probability'].mean()
)
df['snowfall'] = df['snowfall'].fillna(0)
```

```python
df = df.dropna(subset=['latitude', 'longitude'])
```

```python
cols = ['rainfall', 'wind_speed', 'visibility','solar_radiation','precipitation_probability','min_humidity','max_humidity']
(df[cols] < 0).sum()
```

```python
df.loc[df['min_humidity'] < 0, 'min_humidity'] = None
```

```python
(df[cols] < 0).sum()
```

```python
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    print(col, "→ Remaining outliers:", outliers)
```

```python
import folium

# Center map around India (approx)
m = folium.Map(location=[22.5, 78.9], zoom_start=5)
```

```python
for _, row in heatwave_counts.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['heatwave'] / 10,   # size based on heatwave count
        popup=f"Heatwaves: {row['heatwave']}",
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(m)
```
